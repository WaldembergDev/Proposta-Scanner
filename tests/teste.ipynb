```python
# definindo as pendências
def definir_pendencias(os, ano, empresa, pasta_existe, arquivos_existem, revisoes):
  pendencia_encontrada = False
  pendencia_pasta = False if pasta_existe == True else True
  pendencia_solicitacao = False if arquivos_existem[0] == True else True
  pendencia_proposta = False if arquivos_existem[1] == True else True
  # verificando se existem pendência para qualylab ou C&M
  if pendencia_pasta == True:
    pendencia_encontrada = True
  if pendencia_solicitacao == True:
    pendencia_encontrada = True
  if pendencia_proposta == True:
    pendencia_encontrada = True
  if empresa != 'C&M Serviços Ambientais':
    pendencia_aprovacao = False if arquivos_existem[2] == True else True
    pendencia_cond_comerciais = False if arquivos_existem[3] == True else True
    # verificando se existem pendências especificamente para qualylab
    if pendencia_aprovacao == True:
      pendencia_encontrada = True
    if pendencia_cond_comerciais == True:
      pendencia_encontrada = True 
  pendencia_revisoes = revisoes
  if pendencia_revisoes:
    pendencia_encontrada = True
  if not pendencia_encontrada:
    return []    
  if empresa != 'C&M Serviços Ambientais':
    return (
      os,
      ano,
      empresa,
      f'Pendência pasta: {pendencia_pasta}',
      f'Pendência solicitação: {pendencia_solicitacao}',
      f'Pendência Proposta: {pendencia_proposta}',
      f'Pendência Aprovação: {pendencia_aprovacao}',
      f'Pendência Condições Comerciais: {pendencia_cond_comerciais}',
      f'Pendência Revisões: {pendencia_revisoes}'
      )
  else:
   return (
     os,
     ano,
     empresa,
     f'Pendência pasta: {pendencia_pasta}',
     f'Pendência solicitação: {pendencia_solicitacao}',
     f'Pendência Proposta: {pendencia_proposta}',
     f'Pendência revisões: {pendencia_revisoes}'
     )
  

def gerar_dataframe(lista_pendencia: list):
  lista = []
  for i, pendencia in enumerate(lista_pendencia):
    for f in range(3, len(pendencia)):
      if 'True' in pendencia[f]:
        lista.append((pendencia[0], pendencia[1], pendencia[2], pendencia[f]))
  dataframe = pd.DataFrame(lista)
  return dataframe
      

def obter_numeros(texto: str):
  compilado = re.compile(r'\d+')
  resultados = compilado.findall(texto)
  if not resultados:
    return resultados
  # transfornmando todos os valores encontrados em números inteiros
  resultados_convertidos = list(map(int, resultados))
  return resultados_convertidos

# sistema de automação busca de OS
def normalizar_nome(nome: str):
    return unicodedata.normalize('NFKD', nome).encode('ascii', 'ignore').decode('ascii').lower()

# Lista obtida pelo QualiLims
lista_os = [
  ('11601', 2024, 2, 'Comercial 7', 'Aprovado', 'C&M Serviços Ambientais'),
  ('2496', 2024, None, 'Comercial 4', 'VAMO'),
  ('14418', 2024, None, 'Comercial 2', 'AVICOLA CACHOEIRENSE'),
  ('15335', 2024, None, 'Comercial 1', 'NUTRIENTE PRODUTOS E SERVICOS FARMACEUTICOS LTDA')
  ]



# desenvolver uma função que estraia os números da prosta que estiverem após o OS

# arquivos comuns

# arquivos qualylab

# arquivos procurados:
arquivos_procurados = ['solicitacao', 'proposta', 'aprovacao', 'condicoes comerciais']

# buscar o status da proposta

# percorrendo a lista obtida pelo qualilims

# caso qualylab aprovado
lista_pendencia = []
for os in lista_os:
  pasta_existe = False
  arquivos_existem = [False, False, False, False] # solic, prop, aprov, cond comercais
  revisoes = [[False, False] for i in range(os[2])] if os[2] else [] # proposta, solicitacao
  ano = f'OS {os[1]}'
  caminho_ano = caminho / ano
  # caso geral
  for pasta in caminho_ano.iterdir():
    OSs_pasta = obter_numeros(pasta.name)
    if int(os[0]) in OSs_pasta:
      pasta_existe = True # Pasta existe
      for subpasta in pasta.iterdir():
        nome_ajustado = normalizar_nome(subpasta.name)
        # verificando se os documentos existem na pasta do cliente
        for i, arquivo in enumerate(arquivos_procurados):
          if arquivo in nome_ajustado:
            arquivos_existem[i] = True
        # verificando se os documentores referentes a revisão encontram-se na pasta
        for i, valor in enumerate(revisoes):
          proposta_revisao = f'proposta rev{i+1}'
          solicitacao_revisao = f'solicitacao rev{i+1}'
          if proposta_revisao in nome_ajustado:
            revisoes[i][0] = True
          if solicitacao_revisao in nome_ajustado:
            revisoes[i][1] = True
      break
  pendencia = definir_pendencias(os[0], os[1], os[5], pasta_existe, arquivos_existem, revisoes)
  if pendencia:
    lista_pendencia.append(pendencia)

print(lista_pendencia)
```
